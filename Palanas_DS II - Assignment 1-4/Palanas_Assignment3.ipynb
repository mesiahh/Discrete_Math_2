$\textbf{VIII. Handshake Theorem}$

The Handshake Theorem asserts that for any finite undirected graph, the total sum of the degrees of all vertices is equal to twice the number of edges in the graph.

Mathematically, this can be expressed as:

\begin{gather}
\Sigma Degree(v)=2e
\end{gather} 


Example from graph used in finding the degree of a vertex:

\begin{gather}
Edges(e) = 6
\\
TotalDegrees(tv) = 12
\end{gather} 


Applying the Handshake Theorem:

$\therefore$ $2(6)=12$

$\textbf{IX. Travelling in a Graph}$

Traveling in a graph involves moving from one vertex to another by following a sequence of adjacent edges. The graph created below utilized various methods of traversing the graph.

```python
import matplotlib.pyplot as plt
import networkx as nx

G = nx.Graph()

G.add_nodes_from("ABCDEFG")

G.add_edges_from(["AB", "AC", "BC", "BD", "CE", "EF", "DF", "FG"])

pos = {"A": (2,8.1),"B": (5,7.6),"C": (2,6),"D": (4,5.3), "E": (1.3,4.2),"F": (5,3.5),"G": (3,2)}

nx.draw(G, pos= pos, with_labels = True, node_color= "c", node_size = 900, font_color = "black", font_size = 10, font_family = "Times New Roman", font_weight = "bold", width = 1)
plt.title("Travelling in a Graph  \n -Palanas")
plt.margins(0.2)
plt.show()
```

**Walk:** *finite alternative sequence of adjacent vertices and edges*

$A \to G: A\to B \to C \to E \to F \to E \to F \to G$

**Trail:** *walk that does not repeat edges*

$A \to B: A \to B \to C \to E \to F \to D \to B$

**Path:** *does not repeat the same vertex*

$A \to C: A \to B \to D \to F \to E \to C$

**Closed Walk:** *starts and ends in the same vertex*

$A \to A: A \to B \to C \to A$

**Circuit:** *closed walk but does not repeat edges*

$A \to A: A \to B \to C \to A$

**Simple Circuit:** *does not repeat except first and last vertex*

$F \to F: F \to D \to B \to C \to E \to F$

$\textbf{XII. Euler's Circuit}$

One is deemed a Euler's Circuit **if** all vertices have an even degree. The graph implemented below satisfies that condition:

```python
import matplotlib.pyplot as plt
import networkx as nx


G = nx.Graph()

G.add_nodes_from("ABCDEF")

G.add_edges_from(["AB", "AC", "CD", "CB", "BD", "BF", "FE", "CE"])

euler_circuit = list(nx.eulerian_circuit(G))

pos = {"A": (2, 8.1), "B": (5, 7.6), "C": (2, 6), "D": (4, 5.3), "E": (1.3, 4.2), "F": (8, 6)}


nx.draw(G, pos=pos, with_labels=True, node_color="y", node_size=900, font_color="black", font_size=10, font_family="Times New Roman", font_weight="bold", width=1)

for u, v in euler_circuit:
    x0, y0 = pos[u]
    x1, y1 = pos[v]
    dx = x1 - x0
    dy = y1 - y0
    mid_x = (x0 + x1) / 2
    mid_y = (y0 + y1) / 2
    arrow_length_factor = 0.3
    plt.arrow(mid_x, mid_y, dx * arrow_length_factor, dy * arrow_length_factor, head_width=0.1, head_length=0.06, fc='b', ec='b')

plt.title("Euler's Circuit \n -Palanas")
plt.margins(0.2)
plt.show()
```

$\textbf{XIV. Hamiltonian Circuits}$

The graph implemented below is a Hamiltonian circuit wherein the path visits every vertex once and does not repeat same vertex.

```python
import matplotlib.pyplot as plt
import networkx as nx
from matplotlib.patches import FancyArrowPatch

G = nx.Graph()
G.add_nodes_from("ABCDEF")


G.add_edges_from([
    ("A", "B"), ("A", "C"), ("A", "D"),
    ("B", "C"), ("B", "E"),
    ("C", "D"), ("C", "F"),
    ("D", "F"),
    ("E", "F")
])

pos = {"A": (1, 2), "B": (2, 4), "C": (3, 4), "D": (4, 2), "E": (2, 0), "F": (4, 0)}
nx.draw(G, pos, with_labels=True, node_color='skyblue', node_size=900, font_size=12, font_weight='bold')

hamiltonian_circuit = ["A", "B", "C", "F", "D", "A"]

def draw_hamiltonian_edge(pos, p1, p2, color='red', lw=2):
    src = pos[p1]
    tgt = pos[p2]
    
    arrow = FancyArrowPatch(src, tgt, arrowstyle='-|>', color=color, lw=lw, mutation_scale=15)
    plt.gca().add_patch(arrow)

for i in range(len(hamiltonian_circuit) - 1):
    draw_hamiltonian_edge(pos, hamiltonian_circuit[i], hamiltonian_circuit[i + 1])
draw_hamiltonian_edge(pos, hamiltonian_circuit[-1], hamiltonian_circuit[0])

plt.title("Hamiltonian Circuit \n - Palanas")
plt.margins(0.2)
plt.show()
```

$\textbf{XV. Travelling Salesman Problem}$


```python
import matplotlib.pyplot as plt
import networkx as nx

G=nx.Graph()

G.graph["Name"] = "My Graph"

G.add_nodes_from("ABCDE")
G.add_edges_from([
        ("A", "B",{"weight":10}),
        ("A", "C",{"weight":20}),
        ("A", "D",{"weight":30}),
        ("A", "E",{"weight":40}),
        ("B", "C",{"weight":50}),
        ("B", "D",{"weight":60}),
        ("B", "E",{"weight":70}),
        ("C", "D",{"weight":80}),
        ("C", "E",{"weight":90}),
        ("D", "E",{"weight":100}),
        ("E", "D",{"weight":5})
         ])



pos = {
    "A": (3,2),
    "B": (1,7),
    "C": (9,7),
    "D": (5,9),
    "E": (7,2)
}


pos_node_attributes = {}
for node,(x,y) in pos.items():
    pos_node_attributes[node]=(x,y - 0.9)



edge_labels = {(u,v):d["weight"] for u,v,d in G.edges(data=True)}
nx.draw(
        G, pos= pos,
         with_labels = True, 
         node_color ="c", 
         edge_color = ['k', 'r', 'k', 'r', 'r', 'r', 'k', 'k', 'k', 'r', 'r'],
         node_size = 900, 
         font_color = "black", 
         font_size = 20, 
         font_family = "Times New Roman", 
         font_weight = "bold", 
         width = 1
         )


nx.draw_networkx_edge_labels(G, pos = pos, edge_labels=edge_labels, label_pos = 0.5)
plt.margins(0.2)
plt.title("Travelling Salesman Problem \n - Palanas")

plt.show()
```
