\begin{gather}
\left( \sum_{k=1}^n |a_k + b_k|^p \right)^{\frac{1}{p}} \leq \left( \sum_{k=1}^n |a_k|^p \right)^{\frac{1}{p}} + \left( \sum_{k=1}^n |b_k|^p \right)^{\frac{1}{p}}
\end{gather}


Let's set $x_k = |a_k|^{\frac{p}{p-1}}$ and $y_k = |b_k|^{\frac{p}{p-1}}$. This implies that $\frac{1}{p} + \frac{1}{q} = 1$, where $q$ is the conjugate exponent of $p$. So, $q = \frac{p}{p-1}$.

Apply Hölder's inequality:


\begin{gather}
\sum_{k=1}^n |x_k y_k| \leq \left( \sum_{k=1}^n |x_k|^p \right)^{1/p} \left( \sum_{k=1}^n |y_k|^q \right)^{1/q} 
\end{gather}

where $q$ is the conjugate exponent of $p$ given by $\frac{1}{p} + \frac{1}{q} = 1$.

Let $x_k = \frac{|a_k|}{\left(\sum_{k=1}^n |a_k|^p \right)^{1/p}}$ and  $y_k = \frac{|b_k|}{\left(\sum_{k=1}^n |b_k|^p \right)^{1/p}}$

Then, $\sum_{k=1}^n |x_k|^p = 1$ and $\sum_{k=1}^n |y_k|^q = 1.$


Apply Hölder's inequality:

\begin{aligned}
\sum_{k=1}^n \left| \frac{a_k}{\left(\sum_{k=1}^n |a_k|^p \right)^{1/p}} + \frac{b_k}{\left(\sum_{k=1}^n |b_k|^p \right)^{1/p}} \right|^p &\leq \left( \sum_{k=1}^n |x_k|^p \right)^{1/p} \left( \sum_{k=1}^n |y_k|^q \right)^{1/q} \\
\left( \sum_{k=1}^n |a_k + b_k|^p \right)^{1/p} &\leq \left( \sum_{k=1}^n |a_k|^p \right)^{1/p} \left( \sum_{k=1}^n |b_k|^p \right)^{1/q}
\end{aligned} \\

Since $\frac{1}{p} + \frac{1}{q} = 1,$ we have $q = \frac{p}{p-1}$

Hence, 

\begin{aligned}
\left( \sum_{k=1}^n |a_k + b_k|^p \right)^{1/p} &\leq \left( \sum_{k=1}^n |a_k|^p \right)^{1/p} \left( \sum_{k=1}^n |b_k|^p \right)^{1 - \frac{1}{p}} \\
&= \left( \sum_{k=1}^n |a_k|^p \right)^{1/p} \left( \sum_{k=1}^n |b_k|^{\frac{p}{p-1}} \right)^{\frac{1}{p-1}} \\
&= \left( \sum_{k=1}^n |a_k|^p \right)^{1/p} \left( \sum_{k=1}^n |b_k|^p \right)^{\frac{1}{p-1}}
\end{aligned} 


Minkowski's Inequality for sums is proven:
\begin{gather}

**4.1**

Prove Sedrakayan's Lemma $\forall u_i, v_i \in \mathbb{R}^+$:
\begin{gather}
\frac{(\Sigma_{i=1}^n u_i)^2}{\Sigma_{i=1}^n v_i}
\leq
\Sigma_{i=1}^n \frac{(u_i)^2}{v_i}
\end{gather}

**Solution:**


To prove Sedrakayan's Lemma, we'll use the Cauchy-Schwarz Inequality, which states that for any sequences of real numbers $a_i$ and $b_i$, the following holds:

$\sum_{i=1}^n a_i b_i)^2 \leq (\sum_{i=1}^n a_i^2)(\sum_{i=1}^n b_i^2$



Apply Cauchy-Schwarz Inequality to $\sqrt{v_i}$ and $\frac{u_i}{\sqrt{v_i}}$:

\begin{gather}
(\sum_{i=1}^n \sqrt{v_i} \cdot \frac{u_i}{\sqrt{v_i}})^2 \leq (\sum_{i=1}^n (\sqrt{v_i})^2)(\sum_{i=1}^n (\frac{u_i}{\sqrt{v_i}})^2)
\end{gather}


Simplify:

\begin{gather}
(\sum_{i=1}^n u_i)^2 \leq (\sum_{i=1}^n v_i)(\sum_{i=1}^n \frac{u_i^2}{v_i})
\end{gather}
        

Divide both sides by $\sum_{i=1}^n v_i$:
        
\begin{gather}
\frac{(\sum_{i=1}^n u_i)^2}{\sum_{i=1}^n v_i} \leq \sum_{i=1}^n \frac{u_i^2}{v_i}
\end{gather}


$\therefore$, Sedrakayan's Lemma is proven.

Prove Sedrakayan's Lemma if $u_i, v_i$ are square roots of an even integers:
\begin{gather}
\frac{(\Sigma_{i=1}^n u_i)^2}{\Sigma_{i=1}^n v_i}
\leq

\end{gather}

**Solution**
Given Sedrakayan's Lemma for square roots of even integers, we can utilize the fact that if $u_i, v_i$ are square roots of even integers, then $u_i^2, v_i$ are integers.
        
Since $u_i^2, v_i$ are integers, we can apply Sedrakayan's Lemma directly:
        
\begin{gather}
\frac{(\sum_{i=1}^n u_i^2)}{(\sum_{i=1}^n v_i)} \leq \sum_{i=1}^n \frac{u_i^2}{v_i}
\end{gather}
        
This proves Sedrakayan's Lemma for square roots of even integers.
