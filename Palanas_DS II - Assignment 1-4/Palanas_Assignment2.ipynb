Consider the following experiment:

*We conduct $n$ identical trials.
Each trial's outcome belongs to one of $k$ classes or cells.
The probability of a single trial's outcome falling into cell $i$ is $p_i$, where $i = 1,2,...,k$, and the sum of all probabilities equals $1$.
The trials are independent.
We define random variables of interest $Y_1 , Y_2 , . . . , Y_k$, where $Y_i$ represents the number of trials with outcomes falling into cell $i$. Note that $Y_1 +Y_2 +Y_3 +···+Y_k =n$.
The joint probability distribution $P(y) = \frac{n!}{y_1!y_2!y_3!...y_n!} P_1^{y_1} P_2^{y_2} P_3^{y_3}....P_n^{y_n}$*

$\textbf{Question 1:}$ Demonstrate that the expected value $E[y] = \Sigma y \cdot P(y) = n \cdot p_i$

Step 1: Express the expected value $E[y]$ in terms of $P(y)$:

\begin{gather}
E[y] = \Sigma_{y} , y \cdot P(y)
\end{gather}


Step 2: Substitute the given joint probability distribution for $P(y)$.

\begin{gather}
E[y] = \Sigma_{y} , y \cdot \left(\frac{n!}{y_1!y_2!y_3!...y_k!} \cdot p_1^{y_1} p_2^{y_2} p_3^{y_3}....p_k^{y_k}\right) 
\end{gather}

Step 3: Expand the sum:

\begin{gather}
E[y] = \Sigma_{y} \frac{n!}{y_1!y_2!y_3!...y_k!} \cdot (y_1 \cdot p_1) \cdot (y_2 \cdot p_2) \cdot (y_3 \cdot p_3) \ldots (y_k \cdot p_k)
\end{gather}



Step 4: Rearrange the terms:

\begin{gather}
E[y] = \frac{n!}{y_1!y_2!y_3!...y_k!} \cdot \Sigma_{y} (y_1 \cdot p_1) \cdot (y_2 \cdot p_2) \cdot (y_3 \cdot p_3) \ldots (y_k \cdot p_k)
\end{gather}



Step 5: Notice that the sum is a product of sums:

\begin{gather}
E[y] = \frac{n!}{y_1!y_2!y_3!...y_k!} \cdot ( \Sigma_{y_1} ,y_1 \cdot p_1) \cdot (\Sigma_{y_2} , y_2 \cdot p_2) \cdot ( \Sigma_{y_3} , y_3 \cdot p_3) \ldots ( \Sigma_{y_k} , y_k \cdot p_k)
\end{gather}


Step 6: Each individual sum represents the expected value of a binomial distribution with parameters $n$ and $p_i$:

\begin{gather}
E[Y_i] = n \cdot p_i
\end{gather}



Step 7: Substitute the expected values of each $Y_i$:

\begin{gather}
E[y] = \frac{n!}{y_1!y_2!y_3!...y_k!} \cdot (n \cdot p_1) \cdot (n \cdot p_2) \cdot (n \cdot p_3) \ldots (n \cdot p_k)
\end{gather}


Step 8: Notice that $n!$ and $y_1!y_2!y_3!...y_k!$ cancel out:

\begin{gather}
E[y] = n \cdot p_1 \cdot n \cdot p_2 \cdot n \cdot p_3 \ldots n \cdot p_k
\end{gather}



We simplify the expression by combining terms and recognizing that the coefficient $(y_1!y_2!y_3!...y_k!)$ can be simplified to $n!$, the number of permutations of outcomes.

Step 9: Simplify:

\begin{gather}
E[y] = n (p_1 \cdot p_2 \cdot p_3 \ldots p_k)
\end{gather}


Step 10: Recall that $p_1 + p_2 + p_3 + \ldots + p_k = 1 $:

\begin{gather}
E[y] = n \cdot 1 \\
E[y] = n \cdot p_i
\end{gather}




$\therefore$ the expected value $E[y]$ = $\Sigma y \cdot P(y)$, which equals $n \cdot p_i$.

$\textbf{Question 1:}$

*From the following steps shown above, derive the probability distribution of the experiment.*

The probability distribution of $Y$ follows a binomial distribution, which is calculated using the binomial probability mass function:

\begin{gather}
P(Y = y) = \binom{n}{y} \cdot p^y \cdot q^{n-y}
\end{gather}

$\ni$:

- $P(Y = y)$ represents the probability of observing $y$ successes in $n$ trials.
- $\binom{n}{y}$ denotes the binomial coefficient, representing the number of ways to choose $y$ successes out of $n$ trials.
- $p^y$ represents the probability of having $y$ successes.
- $q^{n-y}$ represents the probability of having $(n - y)$ failures.



$\textbf{Question 2:}$

*Show that the expectation of this probability distribution is $E[y] = \Sigma y \cdotp P(y) = np$.*

Step 1: Define the Expectation

\begin{gather}
E[y] = \Sigma_{y = 0}^n \cdot P(y).
\end{gather}


Step 2: Substitute the Binomial Probability Mass Function

\begin{gather}
E[y] = \Sigma_{y = 0}^n \cdot \binom{n}{y} \cdot p^y \cdot q^{n-y}.
\end{gather}


Step 3: Simplify the Expression

Substitute $q$ with $1-p$ and rearrange terms.

\begin{gather}
E[y] = \Sigma_{y = 0}^n \cdot \frac{n!}{y!(n-y)!} \cdot (p)^y \cdot (1-p)^{(n-y)}.
\end{gather}


Step 4: Factor Out Constants

Since $p$ is constant, it can be factored out of the summation.

\begin{gather}
E[y] = np \cdot \Sigma_{y = 0}^n \cdot \frac{(n-1)!}{(y-1)!(n-y)!} \cdot (p)^{(y-1)} \cdot (1-p)^{(n-y)}.
\end{gather}


Step 5: Rearrange the Terms

\begin{gather}
E[y] = np \cdot \Sigma_{y = 1}^n \cdot \frac{(n-1)!}{(y-1)!(n-y)!} \cdot (p)^{(y-1)} \cdot (1-p)^{(n-y)}.
\end{gather}


Step 6: Apply the Binomial Theorem

Use the binomial theorem to expand $(p + q)^ {n -1}$.

\begin{gather}
(p + q)^ {n-1} = \Sigma_{k = 0}^{n-1} \binom{n-1}{y} \cdot (p)^{(k)} \cdot (q)^{(n-1-k)}
\end{gather}


By comparing terms to the binomial theorem, we find:

\begin{gather}
(p + q)^ {n-1} = \Sigma_{y = 0}^{n} \cdot \frac{(n-1)!}{(y)!(n-y)!} \cdot (p)^{(y)} \cdot (q)^{(n-y)}
\end{gather}

Step 7: Simplify the expression

The expansion of $(p + (1-p))^{n-1}$ is $1^{n-1}$, which equals 1. Therefore, the sum becomes 1.

This simplifies to:


$E[y] = np \cdot 1$, leading to the final simplification of $E[y] = np$.



In conclusion, the expectation of a binomial distribution, $E[y] = np$, confirms that the mean equals the product of the number of trials and the probability of success.

$\textbf{Question 3:}$ *Probability of Recovery*

*TExperience has shown that 30% of all persons afflicted by a certain illness recover. A drug company has developed a new medication. Ten people with the illness were selected at random and received the medication; nine recovered shortly thereafter. Suppose that the medication was absolutely worthless. What is the probability that at least nine of ten receiving the medication will recover?*

Given:

Probability of success (recovery) $p = 0.30$

Number of trials $n = 10$

Number of successes (recoveries) $y = 9$

Using the binomial probability mass function:


Solution:


\begin{gather}
P(Y \geq 9) = P(Y = 9) + P(Y = 10) \\
P(Y \geq 9) = \binom{10}{9} \cdot (0.30)^9 \cdot (1 - 0.30)^{10-9} + \binom{10}{10} \cdot (1 - 0.30)^{10-10} \\
P(Y \geq 9) = 10 \cdot (0.30)^9 \cdot (0.70)^1 + 1 \cdot (0.30)^{10} \cdot (0.70)^0 \\
P(Y \geq 9) = 10 \cdot (0.30)^9 \cdot 0.70 + (0.30)^{10} \\
P(Y \geq 9) \approx 0.000137781 + 0.0000059049 \\
P(Y \geq 9) \approx 0.0001436869%
\end{gather}


Thus, the probability of at least nine out of ten individuals recovering is **approximately $0.014%$.**

Probability of success (a fuse being non-defective) $p = 0.95$

Number of trials $n = 5$

\begin{gather}

\end{gather}



Using the binomial probability mass function:

\begin{gather}

P(Y = 0) = (0.95)^5 \\

P(\text{At least one defective}) \approx 1 - 0.7738 \\
P(\text{At least one defective}) \approx 0.2262 \\
\end{gather}



$\therefore$ the probability of observing at least one defective fuse in a sample of 5 fuses is approximately $0.2262$ or $22.62%$.

**Task 3:** Consider a probility distribution of $p(y) = \frac{\lambda^y}{y!} e^{-\lambda}$.

$\textbf{Question 1:}$ 

Find a general formula for the expected value of this distribution.


Given probability distribution: 


\begin{gather}
P(\text{At least one defective}) = 1 - P(\text{No defective}) \\
P(\text{At least one defective}) = 1 - P(Y = 0)
\end{gather}
p(y) = \frac{\lambda^y}{y!} e^{-\lambda}

Expected value formula: 

\begin{gather}
E[Y] = \Sigma_{y=0} ^ \infty \, y \cdot p(y)
\end{gather}



Substitute the Probability Distribution into the Expected Value Formula

\begin{gather}
E[Y] = \Sigma_{y=0} ^ \infty \, y \cdot \frac{\lambda^y}{y!} e^{-\lambda}
\end{gather}


Factor out Constants

\begin{gather}
E[Y] = e^{-\lambda} \Sigma_{y=0} ^ \infty \, \frac{\lambda^y}{(y - 1)!} 
\end{gather}



Since $y = 0$ makes $\frac{\lambda^y}{(y - 1)!}$ undefined, we shift the index of summation to start from $y = 1$.

\begin{gather}
E[Y] = e^{-\lambda} \Sigma_{y=1} ^ \infty \, \frac{\lambda^y}{(y - 1)!} 
\end{gather}


Rewrite $\frac{\lambda^y}{(y - 1)!}$ as $\frac{\lambda^{(y-1)+ 1}}{(y - 1)!}$

\begin{gather}
E[Y] = e^{-\lambda} \Sigma_{y=1} ^ \infty \,  \frac{\lambda^{(y-1)+ 1}}{(y - 1)!}
\end{gather}


Change the index of summation to start from $y = 0$ to simplify the expression.

\begin{gather}
E[Y] = e^{-\lambda} \Sigma_{y=0} ^ \infty \,  \frac{\lambda^{y+1}}{y!}
\end{gather}


Factor out $\lambda$

\begin{gather}
E[Y] = e^{-\lambda} \lambda \Sigma_{y=0} ^ \infty \, \frac {\lambda^{y}}{y!}
\end{gather}


Recall that $e^{\lambda} =  \Sigma_{y=0} ^ \infty \, \frac {\lambda^{y}}{y!}$

\begin{gather}
E[Y] = e^{-\lambda} \lambda e^{\lambda} 
\end{gather}


Simplify: 
\begin{gather}
E[Y] =  \lambda
\end{gather}


$\therefore$, the expected value of the given distribution is **$E[Y] =  \lambda$**
